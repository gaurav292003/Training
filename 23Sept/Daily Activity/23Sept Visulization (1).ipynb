```python
import pandas as pd
import matplotlib.pyplot as plt
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
df = pd.DataFrame(data)
df
```

```python
total_revenue = df.groupby("Product")["Revenue"].sum().reset_index()
print(total_revenue)
```

```python
avg_unit= df.groupby("Product")["Units_Sold"].mean().reset_index()
print(avg_unit)
```

```python
df_q1 = df[df["Month"].isin(["Jan", "Feb", "Mar"])]

pivot_df = df_q1.pivot(index="Month", columns="Product", values= "Revenue")

pivot_df.plot(kind="line",marker="o")

plt.title("Monthly Income trend jan to mar")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()

```

```python
plt.bar(total_revenue["Product"], total_revenue["Revenue"], color=["blue", "orange", "green"])
plt.title("Total Revenue per product")
plt.xlabel("Product")
plt.ylabel("Total Revenue")
plt.grid(axis="y")
plt.legend(title="Product")
plt.show()
```

```python
revenue_per_product= df.groupby("Product")["Revenue"].sum()

plt.figure(figsize=(6,6))
plt.pie(revenue_per_product, labels= revenue_per_product.index, autopct= '%1.1f%%', startangle=140)
plt.title("Contribution of Each Product to Total revenue")
plt.show()
```
